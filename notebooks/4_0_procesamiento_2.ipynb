# Agregamos imperfecciones

```python
import manchas
width, height = 200, 200

DATA_manchas = DATA_DIR.joinpath('manchas')

for i in range(450):
    generator = manchas.CoffeeStainImageGenerator(width, height,'ellipse', num_stains=2)
    generator.image.save(DATA_manchas.joinpath('ellipse_'+str(i)+'.png'))
    
for i in range(450):
    generator = manchas.CoffeeStainImageGenerator(width, height,'line', num_stains=2)
    generator.image.save(DATA_manchas.joinpath('line_'+str(i)+'.png'))
    
for i in range(450):
    generator = manchas.CoffeeStainImageGenerator(width, height,'mix', num_stains=2)
    generator.image.save(DATA_manchas.joinpath('mix_'+str(i)+'.png'))
```

```python
import desenfoque
import impresiones
import impresiones
tool = impresiones.Impresion()
import brillo
```

```python
files = sorted([f for f in DATA_DIR.joinpath('Process','DATA_SET_ENTRENAMIENTO','CLEAN').iterdir() if f.is_file()])
manchas_lista = sorted([f for f in DATA_manchas.iterdir() if f.is_file()])
```

```python
for file in files:
    ruta_nueva = file.parent.parent.joinpath('NOISY',file.name)
    i = random.choice([1,2,3,4])
    if i == 1:
        imagenss= manchas.hacer_manchas(10,file,manchas_lista)
        imagenss.save(str(ruta_nueva))
    if i == 2:
        tool.direccion(file)
        for i in range(2):
            coor2 = tool.coordenas_random()
            imagen = tool.recortar(coor2)
            imagen2 = tool.hacer_impresiones(imagen,coor2[:2])
        tool.imagen.save(str(ruta_nueva))
    if i==3:
        imagen = desenfoque.desenfoques(str(file))
        imagen.save(str(ruta_nueva))
    if i==4:
        imgg = brillo.modicacion(str(file))
        imgg.save(str(ruta_nueva))
    
```
